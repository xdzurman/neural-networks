# Data Analysis

```python
from src.helpers.dataset_helpers import get_train_valid_test
import src.consts as c
import matplotlib.pyplot as plt
import cv2
```

```python
train_paths, valid_paths, _ = get_train_valid_test(c.DATASET_PATH)
```

```python
print('Training images count: {}\nTesting images count: {}'.format(len(train_paths), len(valid_paths)))
```

All images have the same shape:

```python
for img_path in train_paths + valid_paths:
    img = plt.imread(img_path)
    img_shape = img.shape
    if img_shape != (256, 256, 3):
        print('Different size! {}'.format(img_path))
```

Sample images:

```python
fig = plt.figure(figsize=(15, 15))
for i, img_path in enumerate(train_paths[:9]):
    ax = fig.add_subplot(3,3,i+1)
    img = plt.imread(img_path)
    ax.imshow(img)
plt.show()
```
